```python
import pandas as pd
from prophet import Prophet

# Filter for Actuals and HC only (remove Cost Center filter)
HCActuals = path[path['Category'].str.contains('Actual', case=False, na=False) & 
                 (path['Adj. Account'] == 'HC')].reset_index(drop=True)

# Split the 'Time' column into two new columns
HCActuals[['Time.1', 'Time.2']] = HCActuals['Time'].str.split(' - ', expand=True)

#HCActuals[['CC.1', 'CC.2']] = HCActuals['Cost Center'].str.split(' - ', expand=True)

# Prepare output DataFrame
all_forecasts = []

# Loop through each Cost Center
for cc in HCActuals['Cost Center'].unique():
    cc_df = HCActuals[HCActuals['Cost Center'] == cc].copy()
    if cc_df.empty:
        continue
    cc_df = cc_df.drop(columns=['Time','Time.2', 'Adj. Account','Category'], errors='ignore')
    cc_df['Time.1'] = pd.to_datetime(cc_df['Time.1'], format='%Y.%m') + pd.offsets.MonthEnd(0)
    cc_df = cc_df.rename(columns={"Time.1" : "ds", "LC - Local Currency": "y"})
    cc_df = cc_df[['ds', 'y']].dropna()
    if len(cc_df) < 2:  # Prophet needs at least 2 data points
        continue
    modl = Prophet(interval_width=0.95)
    modl.fit(cc_df)
    # Forecast until end of 2026
    last_date = cc_df['ds'].max()
    months_to_forecast = ((2026 - last_date.year) * 12 + (12 - last_date.month))
    future = modl.make_future_dataframe(periods=months_to_forecast, freq='M')
    forecast = modl.predict(future)
    output = forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].round(1)
    output['Cost Center'] = cc
    # Merge actuals for comparison
    output = pd.merge(output, cc_df, on='ds', how='left')
    all_forecasts.append(output)

# Combine all forecasts
result = pd.concat(all_forecasts, ignore_index=True)
result['Cost Center'] = result['Cost Center'].str.split('-').str[0].str.strip()

result['Integer'] = result['ds'].dt.strftime('%Y.%m')

result = result.rename(columns={"y": "Actual HC","yhat": "Forecast","yhat_lower": "Lower Bound","yhat_upper": "Upper Bound"})
result = result[['Cost Center', 'ds', 'Actual HC', 'Forecast', 'Lower Bound', 'Upper Bound','Integer']]
result.to_excel("c:/Users/p40033f/OneDrive - AholdDelhaize.com/Forecast Accuracy/Data Source Files/Prophet_Forecast_All_CostCenters.xlsx", index=False)

result.head()
```

```python
import pandas as pd

# Read your data (replace with your actual file path or DataFrame)
df = pd.read_excel('c:/Users/p40033f/OneDrive - AholdDelhaize.com/Forecast Accuracy/Data Source Files/HC Hierarchy.xlsx', sheet_name = 'Job Hierarchy')  # or pd.read_excel(...)

# Define mapping rules for Title lvl2 and Title lvl3
def map_title_levels(row):
    title = str(row['Job Title']).lower()
    if 'vp' in title or 'evp' in title or 'svp' in title:
        return pd.Series(['VP', 'VP'])
    if 'director' in title:
        return pd.Series(['Director', 'Director'])
    if 'manager' in title or 'hrbp' in title:
        return pd.Series(['Manager', 'Manager'])
    if 'supervisor' in title or 'supv' in title:
        return pd.Series(['Supervisor', 'Analyst'])
    if 'analyst' in title or 'consultant' in title or 'specialist' in title or 'coordinator' in title or 'product owner' in title or 'counsel' in title or 'paralegal' in title or 'buyer' in title:
        return pd.Series(['Analyst', 'Analyst'])
    if 'admin' in title or 'administrator' in title or 'clerk' in title or 'executive asst' in title or 'admin asst' in title:
        return pd.Series(['Admin', 'Admin'])
    if 'intern' in title or 'co-op' in title:
        return pd.Series(['Intern', 'Intern'])
    if 'it engineer' in title or 'software engineer' in title or 'architect' in title or 'domain architect' in title:
        return pd.Series(['IT/Software Engineer', 'IT/Software Engineer'])
    if 'other' in title:
        return pd.Series(['OTHER', 'OTHER'])
    return pd.Series([row['Title lvl2'], row['Title lvl3']])

# Apply the mapping
df[['Title lvl2', 'Title lvl3']] = df.apply(map_title_levels, axis=1)

# Save or return the revised DataFrame
#df.to_excel('revised_job_levels.csv', index=False)
df.head()
```
